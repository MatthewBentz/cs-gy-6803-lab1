```python
# Group 13
# Original text that will be checked
text = "Was it a rat I saw"

# Empty string used to store the cleaned version of the text
cleaned_text = ""

# Loop through each character in the original text
for char in text:

    # Check if the character is a letter or number
    # This removes spaces, punctuation, and symbols
    if char.isalnum():

        # Convert the character to lowercase and add it to cleaned_text
        # This makes the comparison case-insensitive
        cleaned_text += char.lower()

# Reverse the cleaned text using slicing
# [::-1] means start from the end and move backward
reversed_text = cleaned_text[::-1]

# Compare the cleaned text to its reversed version
if cleaned_text == reversed_text:

    # If they are the same, the text is a palindrome
    result = "Palindrome"
else:

    # If they are different, the text is not a palindrome
    result = "Not a Palindrome"

# Print the final result
print(result)
```

```python
# List of inputs to check
inputs = [
    "admin",
    "root;drop table users",
    "hello123",
    "user'='user",
    "safeInput"
]

# Loop through each input in the list
for user_input in inputs:

    # Start by assuming the input is not suspicious
    suspicious = False

    # Check each character in the current input
    for char in user_input:

        # If the character is an apostrophe, semicolon, or equals sign,
        # mark the input as suspicious
        if char in ["'", ";", "="]:
            suspicious = True

    # Print the result for the current input
    if suspicious:
        print(user_input, "- Suspicious input detected")
    else:
        print(user_input, "- Input looks safe")
```

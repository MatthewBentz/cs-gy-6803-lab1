```python
# Group 13

text = "Was it a rat I saw"

cleaned_text = ""

for char in text:
    if char.isalnum():           
        cleaned_text += char.lower()

reversed_text = cleaned_text[::-1]

if cleaned_text == reversed_text:
    result = "Palindrome"
else:
    result = "Not a Palindrome"

print(result)
```

## Assignment: Implementing the RSA Algorithm

---

```python
def rsa_encrypt(message: int, e: int, n: int) -> int:
    return pow(message, e, n)

def rsa_decrypt(message: int, d: int, n: int) -> int:
    return pow(message, d, n)

test_set = {
    "A": {
        "p": 3,
        "q": 11,
        "n": 33,
        "e": 3,
        "d": 7,
        "M": 15,
        "C": 9
    },
    "B": {
        "p": 5,
        "q": 11,
        "n": 55,
        "e": 3,
        "d": 27,
        "M": 42,
        "C": 3
    },
    "C": {
        "p": 7,
        "q": 13,
        "n": 91,
        "e": 5,
        "d": 29,
        "M": 10,
        "C": 82
    }
}

for key, params in test_set.items():
    encrypted = rsa_encrypt(params["M"], params["e"], params["n"])
    decrypted = rsa_decrypt(encrypted, params["d"], params["n"])
    
    print(f"Test {key}:")
    print(f"\tEncrypted: {encrypted} (Expected: {params['C']})")
    print(f"\tDecrypted: {decrypted} (Expected: {params['M']})\n")
```

## Basic Cybersecurity

---

Please explain the following terms.

1. Authentication - the process of verifying the identity of a user or system. It ensures that the person accessing a system is who they claim to be such as logging in with a username and password
2. Authorization - determines what an authenticated user is allowed to do. It controls access levels, such as whether a user can view, edit, or delete information.
3. Confidentiality - ensures that sensitive information is only accessible to authorized users. It protects data from unauthorized access or disclosure
4. Encryption - the process of converting readable data (plaintext) into an unreadable format (ciphertext) using an algorithm and a key, to protect the data from unauthorized access
5. Cipher text - data that has been transformed from readable plaintext into an unreadable format using encryption, making it secure from unauthorized access unless decrypted with the proper key.
6. Hashing - is the process of converting data into a fixed-length string (hash value) using a mathematical algorithm. It is commonly used to store passwords securely
7. Double Strength Encryption - refers to encrypting data twice using encryption algorithms or keys to increase security and make it harder to break
8. Digital Signature - a cryptographic method used to verify the authenticity and integrity of a message or document. It ensures that the data has not been altered and confirms the identity of the sender.

## Hashing

---

What is hash function in cryptography?

### Answer:
A hash function in cryptography is a mathematical algorithim that takes input data and produces a fixed-length string of characters called a hash

Write a hash function in Python to create the hash value of below message:

message = "Information Systems Security Engineering and Management Summer 2026"

To hash the above function use the Python library hashlib (SHA256)

```python
import hashlib

def generate_hash(data):
    data_in_bytes = data.encode('utf-8')
    sha256_hash = hashlib.sha256()
    sha256_hash.update(data_in_bytes)
    return sha256_hash.hexdigest()

message = "Information Systems Security Engineering and Management Summer 2026"
hashed_message = generate_hash(message)

print(f"Raw Message: {message}")
print(f"Hashed Message: {hashed_message}")
```

Hash verification:

Data can be compared to a hash value to confirm its integrity.

The data is hashed at a certain time and the hash value is protected in some way (for this lab, you can note it
down). Later, the data can be hashed again and compared to the protected value. If the hash values match, the
data has not been altered. If the values do not match, the data has been corrupted.

Write a Python function which implements hash_verification(data, original_hash) as explained above. Use that
function to verify the data integrity of the message in the beginning of this section.

```python
def hash_verification(data, original_hash):
    current_hash = generate_hash(data)
    return current_hash == original_hash

print(hash_verification(message, hashed_message))
```
